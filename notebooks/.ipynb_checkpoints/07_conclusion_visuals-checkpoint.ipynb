# 3. Prikaz 3D

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401


pred_svd_df = pd.DataFrame(pred_svd, columns=["userId", "movieId", "true_r", "est", "dummy"])

sample_size = 2000  
sample = test_df.sample(n=min(sample_size, len(test_df)), random_state=42).copy()

# mapiranje userId -> indeks (0..U-1)
unique_users = sorted(sample["userId"].unique())
user_index = {u: i for i, u in enumerate(unique_users)}

# mapiranje movieId -> indeks (0..M-1)
unique_movies = sorted(sample["movieId"].unique())
movie_index = {m: i for i, m in enumerate(unique_movies)}
```

```python
# 1) Stvarne ocene u 3D
sample["user_idx"] = sample["userId"].map(user_index)
sample["movie_idx"] = sample["movieId"].map(movie_index)

X_true = sample["movie_idx"].values  # x = film
Y_true = sample["user_idx"].values   # y = korisnik
Z_true = sample["rating"].values     # z = stvarni rating

# 2) Predviđene ocene (SVD) za iste (user, movie) parove
sample_with_pred = sample.merge(
    pred_svd_df[["userId", "movieId", "est"]],
    on=["userId", "movieId"],
    how="left"
)

X_pred = sample_with_pred["movie_idx"].values
Y_pred = sample_with_pred["user_idx"].values
Z_pred = sample_with_pred["est"].values  # z = predviđeni rating
```

```python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection="3d")

# stvarne ocene
ax.scatter(
    X_true,
    Y_true,
    Z_true,
    marker="o",
    alpha=0.7,
    label="Stvarne ocene"
)

# predviđene ocene (SVD)
ax.scatter(
    X_pred,
    Y_pred,
    Z_pred,
    marker="^",
    alpha=0.5,
    label="Predviđene ocene (SVD)"
)

ax.set_xlabel("Film (indeks)")
ax.set_ylabel("Korisnik (indeks)")
ax.set_zlabel("Rating")

ax.set_title("3D prikaz: korisnici × filmovi × rating (stvarni vs predviđeni)")
ax.legend()

# malo promeni ugao kamere da izgleda lepše, po želji
ax.view_init(elev=20, azim=40)

plt.tight_layout()
plt.show()
```
